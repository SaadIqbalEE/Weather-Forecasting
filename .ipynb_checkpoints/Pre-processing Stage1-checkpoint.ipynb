```python
# classification of "Interval Rain" in 3 Intervals 
#A = [-inf:0] No Rainfall
#B = [0:1] Rainfall
#C = [1:inf] Heavy Rainfall


# Before applying SMOT to reduce the imbalance in the Dataset, Interval rain will be binned in 3 intervals.
# Define bin edges
Rain_labels = [0,1,2] # new labels to be assigned
Rain_cut_bins = [-10, 0, 1, 4] #cutting intervals/criteria [minvalue,0],(0,maxvalue]

# Use pd.cut() to bin the values in the column.
categorized_Rain = pd.cut(df['Interval Rain'], bins=Rain_cut_bins)

categorized_Rain = pd.cut(df['Interval Rain'],\
                                         bins=Rain_cut_bins,labels=Rain_labels,include_lowest=True)
df.insert(11,'categorized Rain',categorized_Rain)

```

```python
#Features & Labels
X = df[['Air Temperature', 'Wet Bulb Temperature', 'Humidity', 'Precipitation Type',\
'Wind Direction', 'Wind Speed', 'Maximum Wind Speed', 'Barometric Pressure',\
'Solar Radiation', 'Heading']].to_numpy()
y = df['categorized Rain'].to_numpy()
```

```python
print("Features matrix: ",X.shape)
print("Label Vector: ", y.shape)
```

```python
#applying Smote to minimize imbalance
from imblearn.over_sampling import SMOTE

# Define the SMOTE object
smote = SMOTE(random_state=91)

# Fit and resample the training data
X_train_res, y_train_res = smote.fit_resample(X, y)
```

```python
# visualize the distribution of categorized Rain
#fig, axes = plt.subplots(1, 2, figsize=(10,4))
plt.hist(y_train_res, bins=3, edgecolor='black')
plt.xlabel("categories (classes)")
plt.ylabel("number of datapoints")
plt.title("Distribution of categorized Rain in evenly distributed after SMOT")
```

```python
from sklearn.model_selection import train_test_split

# Assuming X is the feature set and y is the target variable
print(X_train_res.shape)
print(y_train_res.shape)
```

```python
data = pd.DataFrame(X_train_res)
data['target'] = y_train_res

data_sorted = data.sort_values(by='target').reset_index(drop=True)

X_sorted = data_sorted.drop('target', axis=1).values.reshape(-1, 10)
y_sorted = data_sorted['target'].values

print(X_sorted.shape)
print(y_sorted.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_sorted, y_sorted, test_size=0.3, random_state=52)
```

```python
logi = LogisticRegression(solver='liblinear')
logi.fit(X_train, y_train)
```

```python
y_pred = logi.predict(X_test)
```

```python
def generate_confusion_matrix(y_true, y_pred):
    # visualize the confusion matrix
    ax = plt.subplot()
    c_mat = confusion_matrix(y_true, y_pred)
    sns.heatmap(c_mat, annot=True, fmt='g', ax=ax)

    ax.set_xlabel('Predicted labels', fontsize=15)
    ax.set_ylabel('True labels', fontsize=15)
    ax.set_title('Confusion Matrix', fontsize=15)

y_pred = logi.predict(X_test)
generate_confusion_matrix(y_test, y_pred)
plt.show()
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("accuracy of classification : ", accuracy)
```
